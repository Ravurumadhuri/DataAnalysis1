```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset('tips')
sns.scatterplot(x='total_bill',y='tip',data=data)
plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset('flights')
sns.lineplot(x='month',y='passengers',data=data)
plt.title('Line Plot: Number of Passengers Over Time')
plt.show()
```

```python
data=sns.load_dataset('tips')
sns.boxplot(x='day',y='tip',data=data)
plt.title('Box Plot: Distribution of Tip by Day')
plt.show()
```

```python
data=sns.load_dataset('flights')
pivot_data=data.pivot(values='passengers',index='month',columns='year')
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_data, annot=True, fmt='.0f', cmap='YlGnBu', annot_kws={'size': 8}, linewidths=0.5, linecolor='white')
plt.title('Heatmap: Passengers per Month and Year', fontsize=14)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns




df=pd.read_csv('/content/zomato.csv',encoding='latin-1')
df.head(5)

```

```python
df.shape
print(df.columns)
```

```python
find_null_columns=[features for features in df.columns if df[features].isnull().sum()>0]

find_null_columns
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
df_country.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
final_df.head(2)
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts()
country_names
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color')
plt.show()
```
